### Label Pie chart

```python
fig, ax = plt.subplots(dpi=100)
recipe = ["375 g Flour", "75 g Suger", "250 g Butter", "300 g Berries"]

data = [float(x.split(' ')[0]) for x in recipe]
ingredients = [(x.split(' ')[-1]) for x in recipe]

def func(pct, vals):
    absolute = int(pct/100*np.sum(vals))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)

ax.pie(data, autopct=lambda pct: func(pct, data));
```

```python
fig, ax = plt.subplots(dpi=100)
recipe = ["375 g Flour", "75 g Suger", "250 g Butter", "300 g Berries"]

data = [float(x.split(' ')[0]) for x in recipe]
ingredients = [(x.split(' ')[-1]) for x in recipe]

def func(pct, vals):
    absolute = int(pct/100*np.sum(vals))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)

wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data), 
       textprops=dict(color="w"));

ax.legend(wedges, ingredients, 
          title = 'Ingredients', loc='center left',
          bbox_to_anchor = (1,0, 1,1)) # (x, y, width, height) that the legend is placed in
```

## Nested Pie Chart

```python
fig, ax = plt.subplots(dpi=100)
size = 0.3

vals = np.array([[60,32], [35,20], [26,36]])
vals_sum = vals.sum(axis=1)
vals_flat = vals.flatten()

print(vals_flat, vals_sum)

cmap = plt.get_cmap('tab20c')
outer_color = cmap(np.arange(3)*3)
inner_color = cmap(np.arange(6)*12)
# radius =1 whole 
ax.pie(vals_sum, radius=1, colors=outer_color, 
       wedgeprops=dict(width=size, edgecolor='orange'))

ax.pie(vals_flat, radius=1-size, colors=inner_color, 
       wedgeprops=dict(width=size, edgecolor='orange'))

plt.show()
```
